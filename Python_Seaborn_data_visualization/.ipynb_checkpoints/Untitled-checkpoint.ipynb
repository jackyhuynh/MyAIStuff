<a href="https://numpy.org/doc/stable/reference/generated/numpy.linspace.html">linspace</a>

```python
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
# col1 is create using linspace wiyh 10 vars
col1 = np.linspace(0, 10, 1000)

# col2 is a collection of col1 sin vars  
col2 = np.sin(col1) 

# DataFrame({ColName1:Vale1, ColName2:Value2,...}) function create dataframe 
df = pd.DataFrame({'C1':col1, 'C2':col2})

df.head(10)
```

```python
# using matplotlib and seaborn 
plt.style.use('seaborn-darkgrid')
%matplotlib inline
sns.lineplot(x=df.C1, y=df.C2, data=df)
plt.show
```

```python
""" - Adjusting background color using axes.facecolor
    - Changing label size using xtick.labelsize, ytick.labelsize"""

plt.figure(figsize=(14,6))
sns.set(rc={"axes.facecolor":"#FFFDE7", "axes.grid":True, 'xtick.labelsize':14, 'ytick.labelsize':6})
sns.lineplot(x=df.C1, y=df.C2, data=df, linewidth=2)
sns.lineplot(x=df.C1, y=df.C2+1, data=df, linewidth=2)
plt.show()
```

```python
""" - Adjusting background color using axes.facecolor
    - Changing label size using xtick.labelsize, ytick.labelsize"""

plt.figure(figsize=(14,6))
sns.set(rc={"axes.facecolor":"#283747", "axes.grid":False, 'xtick.labelsize':14, 'ytick.labelsize':6})
sns.lineplot(x=df.C1, y=df.C2, data=df, color ="#FF5722",linewidth=2)
sns.lineplot(x=df.C1, y=np.cos(df.C1) , color="#FFEB2B", data=df, linewidth=2)
plt.show()
```

```python
""" - Quick demostration on the Canada immigration data set
    - Data Exploration, Visualization, and Tidy"""

canada = pd.read_csv("data/Canada.csv")
canada.head(10)
```

```python
# Explore data
canada.columns
```

```python
""" - Tidy the data
    - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html
"""

# Drop the unecessary, and tidy the data
# inplace = TRUE :inplacebool, default False
# If False, return a copy. Otherwise, do operation inplace and return None.
canada.drop(columns=['Type','Coverage','AREA', 'AreaName', 'REG', 'RegName',
       'DEV', 'DevName', 'Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46',
       'Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'],inplace=True)

canada.head(5)
```

```python
canada.rename(columns={'OdName':'CountryName'},inplace=True)
canada.set_index(canada.CountryName,inplace=True)
canada.head()
```

```python

canada.index.name=None

# delete the Country Name columns
del canada['CountryName']

# drop all the na values
canada.dropna(inplace=True)

# Check
canada.head()
```

```python
# create an instance of the transpose
canada = canada.transpose()
```

```python

canada.head(10)
```

```python
plt.figure(figsize=(16,6))
plt.title("Immigrant from Australia, Pakistan", fontsize = 16)
plt.ylabel("Population",fontsize=14)
sns.set( rc={"axes.facecolor":"#283747","axes.grid":False, 'xtick.labelsize':10,'ytick.labelsize':10})
plt.xticks(rotation=45) # Rotating X ticks by 45 degrees
sns.lineplot(x=canada.index.values, y =canada['Australia'], color='white', label='Australia' )
sns.lineplot(x=canada.index.values, y =canada['Pakistan'], color='green', label='Pakistan' )
sns.lineplot(x=canada.index.values, y =canada['Argentina'], color='yellow', label='Argentina' )

# the legend box modification
plt.legend(facecolor='gray', fontsize='large',edgecolor='black',shadow=True)
plt.show()
```

```python

plt.figure(figsize=(16,6)) # Frame of the data set

plt.title("Canada Immigrations", fontsize=12)

x = canada.index.values
for i in canada.columns:
    if canada[i].name != 'Total' and canada[i].name!='Unknown':
        y=canada[i]
        sns.lineplot(x,y)
        
plt.xlabel("Year")
plt.ylabel("Country")
plt.show()
```

## Histograms, KDE, and densities

Often in statistical data visualization, all you want is to plot histograms and joint distributions of variables. We have seen that this is relatively straightforward in Matplotlib:

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], density=True, alpha=0.5)
```

Rather than a histogram, we can get a smooth estimate of the distribution using a kernel density estimation, which Seaborn does with sns.kdeplot:

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
sns.distplot(data['x'])
sns.distplot(data['y']);
```

If we pass the full two-dimensional dataset to kdeplot, we will get a two-dimensional visualization of the data:

```python
sns.kdeplot(data);
```
